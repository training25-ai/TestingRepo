```python
# locating a record
# locating a specific record by pokemon
Charizard_record = df.loc[df['pokemon'] == 'Charizard']
print(Charizard_record)

# locating multiple records
# locating based on filtering a value
high_rating = df.loc[df['hp'] > 70]
print(high_rating)

# locate using multiple conditions
starter_pack = df.loc[(df['pokemon'] == 'Pikachu') | (df['hp'] < 40)]
print(starter_pack)

# locate by index
seventh_row = df.iloc[8]
print(seventh_row)
```
